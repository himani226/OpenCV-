# Detect the blue color

```python
#Read image
mg = cv2.imread('colors.jpg')
new_img = cv2.cvtColor(mg, cv2.COLOR_BGR2RGB)
# Converts images from RGB to HSV
hsv = cv2.cvtColor(new_img, cv2.COLOR_RGB2HSV)
lower_blue = np.array([110,50,50])
upper_blue = np.array([130,255,255])

# Here we are defining range of bluecolor in HSV
# This creates a mask of blue coloured 
# objects found in the frame.
mask = cv2.inRange(hsv, lower_blue, upper_blue)

# The bitwise and of the frame and mask is done so 
# that only the blue coloured objects are highlighted 
# and stored in res
res = cv2.bitwise_and(new_img, new_img, mask= mask)
plt.imshow(new_img)
```

### Real-time blue color detection using open cv

```python
import cv2
import numpy as np

cap = cv2.VideoCapture(0)
  
while 1:
    ret,frame =cap.read() 
    # Converts images from BGR to HSV
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    lower_blue = np.array([110,50,50])
    upper_blue = np.array([130,255,255])

    # Here we are defining range of bluecolor in HSV
    # This creates a mask of blue coloured 
    # objects found in the frame.
    mask = cv2.inRange(hsv, lower_blue, upper_blue)

    # The bitwise and of the frame and mask is done so 
    # that only the blue coloured objects are highlighted 
    # and stored in res
    res = cv2.bitwise_and(frame, frame, mask= mask)
    cv2.imshow('Original',frame) # to display the original frame
    cv2.imshow('Blue Detector',res) # to display the blue object output
  
    if cv2.waitKey(1)==27:
        break
    # this function will be triggered when the ESC key is pressed
    # and the while loop will terminate and so will the program
cap.release()

# using the wait key function to delay the
# closing of windows till any key is pressed
cv2.destroyAllWindows()
```

**Morphological Operations for Image Processing**

Morphological operations are one of the Image processing techniques that processes image based on shape. This processing strategy is usually performed on binary images. 


It needs two inputs, one is our original image, second one is called structuring element or kernel which decides the nature of operation. Two basic morphological operators are Erosion and Dilation. 

**EROSION**

The basic idea of erosion is just like soil erosion only, it erodes away the boundaries of foreground object. The kernel slides through the image (as in 2D convolution). A pixel in the original image (either 1 or 0) will be considered 1 only if all the pixels under the kernel is 1, otherwise it is eroded (made to zero).

```python
gray_img = cv2.imread('colors.jpg', cv2.IMREAD_GRAYSCALE)

kernel = np.ones((6,6),np.uint8)
er = cv2.erode(gray_img,kernel,iterations=1)

plt.subplot(121),plt.imshow(gray_img, cmap='gray'),plt.title("original")
plt.subplot(122),plt.imshow(er, cmap='gray'),plt.title("Eroded Image")
```

```python
white = cv2.imread("colors.jpg")

kernel = np.ones((3,3),np.uint8)
er = cv2.erode(white,kernel,iterations=1)

plt.subplot(121),plt.imshow(white),plt.title("original")
plt.subplot(122),plt.imshow(er),plt.title("Eroded Image")
```

**DILATION**

It is just opposite of erosion. Here, a pixel element is ‘1’ if atleast one pixel under the kernel is ‘1’. So it increases the white region in the image or size of foreground object increases. Normally, in cases like noise removal, erosion is followed by dilation. Because, erosion removes white noises, but it also shrinks our object. So we dilate it. Since noise is gone, they won’t come back, but our object area increases. It is also useful in joining broken parts of an object.

```python
white = cv2.imread("colors.jpg")

kernel = np.ones((4,4),np.uint8)
dl = cv2.dilate(white,kernel,iterations=1)

plt.subplot(121),plt.imshow(white),plt.title("original")
plt.subplot(122),plt.imshow(dl),plt.title("Dilation Image")
```

**OPENING**

Opening is just another name of erosion followed by dilation. It is useful in removing noise, as we explained above. Here we use the function, cv2.morphologyEx().

```python
white = cv2.imread("colors.jpg")

kernel = np.ones((5,5),np.uint8)
opening = cv2.morphologyEx(white,cv2.MORPH_OPEN,kernel)

plt.subplot(121),plt.imshow(white,cmap='gray'),plt.title("original")
plt.subplot(122),plt.imshow(opening,cmap='gray'),plt.title("Opening Image")
```

**Thresholding**


> Binary Thresholding: Let's say we wish to convert an image into a binary image i.e. assign a pixel either a value of 0 or 1. To do this, we can perform thresholding. For instance, if the Threshold (T) value is 160, then all pixels with values greater than 160 would be assigned a value of 1, and all pixels with values lesser than or equal to that would be assigned a value of 0.


```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

img = cv.imread('images/coins_2.jpg', cv.IMREAD_GRAYSCALE)

ret,thresh1 = cv.threshold(img,110,255,cv.THRESH_BINARY)
ret,thresh2 = cv.threshold(img,110,255,cv.THRESH_BINARY_INV)
ret,thresh3 = cv.threshold(img,120,255,cv.THRESH_TRUNC)
ret,thresh4 = cv.threshold(img,120,255,cv.THRESH_TOZERO)
ret,thresh5 = cv.threshold(img,120,255,cv.THRESH_TOZERO_INV)

titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]
for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray',vmin=0,vmax=255)
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
# find contours
coins_contours, _ = cv2.findContours(coins_binary, 
                                     cv2.RETR_EXTERNAL, 
                                     cv2.CHAIN_APPROX_SIMPLE)

# make copy of image
coins_and_contours = np.copy(coins)

# find contours of large enough area
min_coin_area = 30
large_contours = [cnt for cnt in coins_contours 
                  if cv2.contourArea(cnt) > min_coin_area]

# draw contours
cv2.drawContours(coins_and_contours, 
                 large_contours, -1, (0,0,255))

plt.imshow(cv2.cvtColor(coins_and_contours,
                        cv2.COLOR_BGR2RGB))

# print number of contours
print('Number of coins: ', len(large_contours))
```

```python
# create copy of image to draw bounding boxes
bounding_img = np.copy(coins)

# for each contour find bounding box and draw rectangle
for contour in large_contours:
    x, y, w, h = cv2.boundingRect(contour)
    cv2.rectangle(bounding_img, (x, y), (x + w, y + h),
                  (0, 255, 0), 3)

plt.imshow(cv2.cvtColor(bounding_img, cv2.COLOR_BGR2RGB))
```
